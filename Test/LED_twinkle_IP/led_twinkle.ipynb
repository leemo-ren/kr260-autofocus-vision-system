```python
from pynq.lib import AxiGPIO
led_instance = axi_gpio_design.ip_dict['axi_gpio_led'] # 
led = AxiGPIO(led_instance).channel1
led[0:1].write(0x1)
```

```python
from time import sleep
while(1):
    led[0:1].write(0x01)
    sleep(1)
    led[0:1].write(0x00)
    sleep(1)
```
