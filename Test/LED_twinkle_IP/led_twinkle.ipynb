```python
from pynq import Overlay
axi_gpio_design = Overlay("./overlay/led_twinkle_ip.bit")
axi_gpio_design.ip_dict  # 显示所有的硬件信息
```

```python
hex(axi_gpio_design.ip_dict["axi_gpio_led"]["phys_addr"]) # 显示地址信息，应该与vivado中设置的地址相同，作为验证
```

```python
from pynq.lib import AxiGPIO
led_instance = axi_gpio_design.ip_dict['axi_gpio_led'] # 实例化
led = AxiGPIO(led_instance).channel1
led[0:1].write(0x1)
```

```python
from time import sleep
while(1):
    led[0:1].write(0x01)
    sleep(1)
    led[0:1].write(0x00)
    sleep(1)
```
